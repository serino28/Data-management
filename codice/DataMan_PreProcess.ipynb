```python
import pandas as pd
import numpy as np
```

```python
crypto = pd.read_csv('crypto.csv', sep=',')
```

```python
crypto.head()
```

```python
crypto.columns
```

```python
crypto.columns = ['img', 'name', 'ticker', 'price', '7d(%)', 'sign', 'market_cap($)', 'volume_7d($)', 'total_supply', 'dominance(%)']
```

```python
crypto['tweets'] = np.nan
```

```python
for index, row in crypto.iterrows():
    row['price'] = row['price'].replace('$', '')
    row['price'] = row['price'].replace(',', '')
    row['7d(%)'] = row['7d(%)'].replace('%', '')
    row['market_cap($)'] = row['market_cap($)'].replace('$', '')
    row['market_cap($)'] = row['market_cap($)'].replace(',', '')
    row['market_cap($)'] = row['market_cap($)'].replace(' ', '')
    row['volume_7d($)'] = row['volume_7d($)'].replace('$', '')
    row['volume_7d($)'] = row['volume_7d($)'].replace(',', '')
    row['volume_7d($)'] = row['volume_7d($)'].replace(' ', '')
    row['total_supply'] = row['total_supply'].replace(' '+row['ticker'], '')
    row['total_supply'] = row['total_supply'].replace(',', '')
    row['dominance(%)'] = row['dominance(%)'].replace('%', '')
```

```python
crypto
```

```python
crypto['price'][662] = '0.00000000003249'
crypto['price'][849] = '0.0000000002954'
crypto['price'][1376] = '0.00000000008419'
crypto['price'][1415] = '0.000000000005715'
crypto['price'][1637] = '0.00000000001015'
crypto['price'][1831] = '0.0000000004787'
```

```python
crypto.loc[crypto['sign'].str.contains('up'),'7d(%)'] = pd.to_numeric(crypto[crypto['sign'].str.contains('up')]['7d(%)'])
crypto.loc[crypto['sign'].str.contains('down'),'7d(%)'] = pd.to_numeric('-'+crypto[crypto['sign'].str.contains('down')]['7d(%)'])
```

```python
del crypto['sign']
```

```python
crypto['total_supply'] = pd.to_numeric(crypto['total_supply'])
```
